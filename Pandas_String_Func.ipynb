```python
import pandas as pd

data = {
    'Name': ['ram', 'shyam', 'Radha'],
    'Age': [12, 45, 67],
    'Number': [344567, 5678, 8966]
}

df = pd.DataFrame(data)
df

```

```python
# str.lower
df['Lower_name'] = df['Name'].str.lower()
df
```

```python
df['Upper_name'] = df['Name'].str.upper()
df
```

```python
# str.len()
df['name_len'] = df['Name'].str.len()
df
```

```python
# str.strip()
df['Name'][0]
df['name_strip'] = df['Name'].str.strip()
df

```

```python
# lstrip and rstrip
df['name_lstrip'] = df['Name'].str.lstrip()
df
df['name_rstrip'] = df['Name'].str.rstrip()
df
```

```python
# str.split()
df['Name_split'] = df['Name'].str.split(' ', expand = True)
df
```

```python
# str.contains()
df['Name'].str.contains('am')
df[df['Name'].str.contains('am')]
```

```python
# str.replace()
df['name_replace'] = df['Name'].str.replace('a', 'u')
df
```

```python
# str.startswith() or endswith()
df['Name'].str.startswith('r')
df[df['Name'].str.startswith('r')]

```

```python
# str.cat()
df['Name'].str.cat(df['Name'], sep = ' ')
```

```python
# str.get()
df['Name'].str.get(-1)
```

```python
# str.slice()
df['slice_name'] = df['Name'].str.slice(0,3)
df
```

```python
# str.find() find occurence index, not find return -1
df['Name'].str.find('m')
```

```python
# rfind()
df['Name'].str.rfind('a')
```
