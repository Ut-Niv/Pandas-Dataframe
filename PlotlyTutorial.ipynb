# Histogram using graph_objects class


```python
# plotly.graph_objects.Histogram()

import plotly.express as px
import plotly.graph_objects as go
df = px.data.iris()

df.head()

```

```python
fig = go.Figure(data = [go.Histogram(x = df['sepal_width'])])
fig.show()
```

```python
fig = go.Figure(data = [go.Histogram(y = df['sepal_width'])])
fig.show()
```

```python
# Stacked Histograms

fig = go.Figure()
fig.add_trace(go.Histogram(x = df['sepal_width']))
fig.add_trace(go.Histogram(x = df['species_id']))

fig.show()
fig.update_layout(barmode='stack')
fig.show()

```

# Creating cumulative histogram

```python
import plotly.express as px
import plotly.graph_objects as go

df = px.data.iris()

fig  = go.Figure(data = [go.Histogram(x=df['sepal_width'], cumulative_enabled = True)])
fig.show()

```

```python
import plotly.graph_objects as px
import numpy as np
np.random.seed(42)

random_x= np.random.randint(1,101,100)
random_y= np.random.randint(1,101,100)

plot = px.Figure(data=[px.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers',
     marker=dict(
       color = [10, 20, 30, 50],
       size = [10, 30, 50, 80],
       showscale=True
     )
)])

plot.show()
```

# Gantt Chart

```python
import plotly.figure_factory as ff

df = [
    dict(Task="A", Start='2020-01-01', Finish='2009-02-02'),
    dict(Task="Job B", Start='2020-03-01', Finish='2020-11-11'),
    dict(Task="Job C", Start='2020-08-06', Finish='2020-09-21')
]
fig = ff.create_gantt(df)
fig.show()
```

```python
import plotly.figure_factory as ff

df = [
    dict(Task="A", Start='2020-01-01', Finish='2009-02-02'),
    dict(Task="Job B", Start='2020-03-01', Finish='2020-11-11'),
    dict(Task="Job C", Start='2020-08-06', Finish='2020-09-21')
]

colors = ['#7a0504', (0.2, 0.7, 0.3), 'rgb(210, 60, 180)']
fig = ff.create_gantt(df, colors = colors, show_colorbar = True, show_hover_fill= True)
fig.show()
```

# Contour Plot

```python
import plotly.graph_objects as go
import numpy as np

x = np.arange(0,50,2)
y = np.arange(0,50,3)
[X, Y] = np.meshgrid(x, y)

Z = np.cos(X /2) + np.sin(Y/4)
fig = go.Figure(data=go.Contour(x = x, y=y, z = Z))

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

data = [
     [1,2,3,4,5],
     [3,4,5,6,7],
     [7,8,9,6,4],
     [3,7,2,4,2]
]
fig = go.Figure(data = go.Contour(z = data))
fig.show()
```

```python
fig = go.Figure(data = go.Contour(z = data, colorscale='rainbow'))
fig.show()
```

```python
fig = go.Figure(data = go.Contour(z = data, contours = dict(start = 1, end=5, size=.2)))
fig.show()
```

```python
# Adding Labels on Contour Line
fig = go.Figure(data = go.Contour(z = data, contours = dict(showlabels=True, coloring='lines')))
fig.show()
```

# HeatMaps

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

x = np.arange(0,50,2)
y = np.arange(0,50,3)

[X, Y] = np.meshgrid(x, y)
z = np.cos(X/2) + np.sin(Y/4)

fig = go.Figure(data = go.Heatmap(x = x, y = y, z= z))
fig.show()
```
