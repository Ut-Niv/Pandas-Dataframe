```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('/content/heart.csv')
data.head()




```

```python
data.shape
```

```python
# draw corelation matrix
plt.figure(figsize = (17,6))
sns.heatmap(data.corr(), annot = True)
# parameters inside the box shows using `annot = True`
```

```python
# How many people has heart disease and how many have not ?
data.columns

data['target'].value_counts()
sns.countplot(x = 'target', data =  data)
```

```python
# count of male and female
data.columns
sns.countplot(x = 'sex', data = data)
plt.xticks([0,1], ['Female', 'male'])
plt.show()
```

```python
# find gender distribution according to the target variable
data.columns

sns.countplot(x = 'sex', hue = 'target', data = data)
plt.xticks([0,1], ['Male', 'Female'])
plt.legend(labels = ['No-Disease', 'Disease'])
plt.show()
```

```python
# chek age distribution in dataset
sns.distplot(data['age'], bins = 20)
plt.show()
```

```python
# check chestpain type is more common
sns.countplot(x = 'cp', data = data)
plt.xticks([0,1,2,3], ['typical angina', 'atypical angina', 'non-anginal pain', 'asymptomatic'])
plt.show()
```

```python
# show chest pain distribution As per taget variable
data.columns

sns.countplot(x = 'cp', hue = 'target', data = data)
plt.legend(labels = ['No-Disease', 'Disease'])
plt.show()
```

```python
# show fasting blood sugar distribution according to target variable
data.columns

sns.countplot(x = 'fbs', hue = 'target', data = data)
plt.legend(labels = ['No-Disease', 'Disease'])
plt.show()

```

```python
# Check resting blood pressure distribution
data.columns

data['trestbps'].hist()
```

```python
# compare resting blood pressure as per sex column

temp = sns.FacetGrid(data, hue = 'sex', aspect = 4)
temp.map(sns.kdeplot,'trestbps', fill = True)
plt.legend(labels = ['Male', 'Female'])
plt.show()
```

```python
# plot continues values
data.columns

val1 = []
val2 = []

for col in data.columns:
  if data[col].nunique() <=10:
    val1.append(col)
  else:
    val2.append(col)

```
