```python
# Find aggregrate sale among branches (categorial columns)

cat_val = []
num_val = []


for column in data.columns:
  if data[column].nunique() > 10:
    num_val.append(column)
  else:
    cat_val.append(column)

```

```python
# Find aggregrate sale among branches (categorial columns)
sns.countplot(x = 'Branch', data = data)

```

```python
# Bivariate Analysis / Multivariate analysis
```

```python
# Does the Cost of goods sold affect the ratings that the customers provide?
sns.scatterplot(x ='cogs', y = 'Rating', data = data)
```

```python
sns.scatterplot(x ='gross income', y = 'Rating', data = data)
```

```python
# find the most profitable branch as per gross income (numerical - cateorial)
sns.barplot(x ='Branch', y= 'gross income', data = data)
```

```python
# find the product line that generates the most income(numerical - categorial)
data.columns
sns.barplot(x = 'Product line', y = 'gross income', data = data)
plt.xticks(rotation = 60)
plt.show()
```

```python
# find the highest unit price in the product line
sns.barplot(x = 'Product line', y = 'Unit price', data = data)
plt.xticks(rotation = 60)
plt.show()
```
