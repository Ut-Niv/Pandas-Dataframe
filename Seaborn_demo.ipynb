# Histogram PLot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = sns.load_dataset('penguins')
df.head()
```

```python
# use histplot instead of distlot
sns.histplot(df['bill_length_mm'])
plt.show()
```

```python
sns.histplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75])
plt.show()
```

```python
sns.histplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75], kde = True)
plt.show()
```

```python
# rug not worked with histplot
sns.distplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75], kde = True, rug = True)
plt.show()
```

```python
sns.histplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75], color= 'g')
plt.show()
```

# Scatter Plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('penguins')
df.head()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df)
plt.show()

```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df, hue = 'sex')
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,hue = 'sex', style= 'sex')
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,
                hue = 'sex', style= 'sex', size = 'sex', sizes = (100,40))
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,\
                hue = 'sex', style= 'sex', palette = 'Accent')
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,
                hue = 'sex', style= 'sex', palette = 'Accent', alpha = 0.5)
```

```python
# marker parameter
m = {'Male': "*", 'Female': 'o'}
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,
                hue = 'sex', style= 'sex', palette = 'Accent', markers = m)
```

# Voilin Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('tips')
data.head()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data)
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data, hue='time')
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data, hue='time', linewidth = 5)
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data, hue='time', palette = 'Dark2')
plt.show()
```

```python
sns.violinplot(x='time', y = 'total_bill', data = data, hue='time', order=['Dinner', 'Lunch'])
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data, hue='time', palette = 'Dark2', saturation = 0.4)
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data, hue='time', color='y')
plt.show()
```

```python
sns.violinplot(x='day', y = 'total_bill', data = data, hue='sex', split = True, scale = "count")
plt.show()
```

```python
# single violin plot
sns.violinplot(y = data['total_bill'])
plt.show()
```

## Pair Plot

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
plt.show()
```

```python
sns.pairplot(df, vars = ['total_bill', 'tip'], hue='sex')
```

```python
sns.pairplot(df, hue='sex',hue_order = ['Female', 'Male'])
```

```python
sns.pairplot(df,hue='sex',hue_order = ['Female', 'Male'], palette = 'Dark2')
plt.show()
```

```python
sns.pairplot(df,hue='sex',hue_order = ['Female', 'Male'], palette = 'Dark2', kind = 'kde')
plt.show()
```

```python
sns.pairplot(df,hue='sex',hue_order = ['Female', 'Male'], palette = 'Dark2', kind = 'kde', diag_kind='hist')
plt.show()
```

```python
sns.pairplot(df,hue='sex',hue_order = ['Female', 'Male'], markers=['>', '*'])
plt.show()
```
