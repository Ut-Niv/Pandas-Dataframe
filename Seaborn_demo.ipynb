# Histogram PLot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = sns.load_dataset('penguins')
df.head()
```

```python
# use histplot instead of distlot
sns.histplot(df['bill_length_mm'])
plt.show()
```

```python
sns.histplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75])
plt.show()
```

```python
sns.histplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75], kde = True)
plt.show()
```

```python
# rug not worked with histplot
sns.distplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75], kde = True, rug = True)
plt.show()
```

```python
sns.histplot(df['bill_length_mm'], bins = [35, 40, 45, 50, 55, 60, 65, 70, 75], color= 'g')
plt.show()
```

# Scatter Plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('penguins')
df.head()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df)
plt.show()

```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df, hue = 'sex')
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,hue = 'sex', style= 'sex')
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,
                hue = 'sex', style= 'sex', size = 'sex', sizes = (100,40))
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,\
                hue = 'sex', style= 'sex', palette = 'Accent')
plt.show()
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,
                hue = 'sex', style= 'sex', palette = 'Accent', alpha = 0.5)
```

```python
# marker parameter
m = {'Male': "*", 'Female': 'o'}
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df,
                hue = 'sex', style= 'sex', palette = 'Accent', markers = m)
```
