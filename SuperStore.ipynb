```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

pd.set_option('display.float_format', lambda x: '%.3f' % x)
warnings.filterwarnings('ignore')
```

```python
# Gather and clean data
data = pd.read_csv('/content/superstore_dataset2011-2015.csv', encoding = 'ISO-8859-1')
data.head()
```

```python
# drop unnecessary columns
data.columns

data = data.drop(['Row ID', 'Order ID', 'Customer ID', 'Postal Code'], axis = 1)
```

```python
data.columns
```

```python
### Technology products have the highest profit margin compared to other product categories
```

```python
catProfit = data.groupby('Category')['Profit'].sum()
catProfit
```

```python
catProfit.plot(kind = 'bar')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()
```

```python
# the east region has the highest sales compared to other region
res_sales = data.groupby('Region')['Sales'].sum()
res_sales

res_sales.plot(kind = 'bar')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# sales are higher during certain months of the year
data['Order Month'] = pd.DatetimeIndex(data['Order Date']).month

res = data.groupby('Order Month')['Sales'].sum()
res.plot(kind = 'line')
plt.xlabel('Month')
plt.ylabel('Total sells')
plt.show()
```
