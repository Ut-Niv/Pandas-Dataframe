```python
import pandas as pd
list = {
    'Name': ['Ram', 'Shyam', 'Anubhav'],
    'Gender': ['M', 'M', 'F'],
    'Score': [23, 34, 45]
}

data = pd.DataFrame(list)
data
```

```python
# values_count
data['Gender'].value_counts()
data['Gender'].value_counts(normalize = True)
```

```python
# where()
data.where(data['Score'] > 25, other = 0)
data.where(data['Score'] > 25, other = 0).all(1)
data[data.where(data['Score'] > 25, other = 0).all(1)]
```

```python
# isin method
data['Score'].isin([34,45])
```

```python
data[data[['Score', 'Gender']].isin({
    'Score': [34,45],
    'Gender': ['M']
})].all(1)
```

```python
# groupby
group = data.groupby('Gender')
group
```

```python
group['Score'].mean()
```

```python
group['Score'].agg(['mean', 'sum', 'count'])
```

```python
# pivot_table
data
```

```python
data.pivot_table(index= 'Gender', values = 'Score', aggfunc = 'mean')
```

```python
data.pivot_table(index=[ 'Gender', 'Name'], values = 'Score', aggfunc = 'mean')
```

```python
# nlargest and nsmallest
data['Score'].nlargest(2)
```

```python
data.nsmallest(3, 'Score')
```

```python
# query
data.query('Score > 40 and Gender == "F"')
```

```python
# apply
data['Gender_lower'] = data['Gender'].apply(str.lower)
```
